### Import libraries 

```python
import pandas as pd
```

### Load the data

```python
df = pd.read_csv('full_data.csv')
df
```

### Remove rows where activityID = 0

```python
df = df[df['activityID'] != 0]

# check for all unique 'activityID' values
print(df.shape)
df['activityID'].unique()
```

### Handle missing values

##### Check how much missing values in every column

```python
df.isnull().sum()
```

As we can see, the biggest number of missing values are in 'heart rate'. <br>
So we will fill them with the most appropriate values while the other values we will fill be mean values of the columns

##### Check the number of missing values in 'heart rate' for each 'activityID' value 

```python
actId_group = df.groupby('activityID')['heart rate']
missing_values_group = actId_group.apply(lambda x: x.isnull().sum())

# add a column with total count of 'ActivityID' values 
missing_values_group = pd.concat([missing_values_group, df['activityID'].value_counts()], axis=1)

# Set correct column names
missing_values_group.index.name = 'ActivityID value'
missing_values_group.columns = ['NaN count in heart rate', 'ActivityID total count']
missing_values_group
```

As we can see, 'heart race' missing values distributed among all activities in equal proportions. <br>
So the best reason is replace missing values with corresponding ActivityID's mean values

```python
df['heart rate'].fillna(actId_group.transform('mean'), inplace=True)

# check if all work correct 
df.isnull().sum()
```

As we can see, now 'heart rate' has not any missing values

##### Replace remaining missing values with mean values of the corresponding column 

```python
df = df.apply(lambda x: x.fillna(x.mean()))

# check current count of missing values
df.isnull().sum()
```

##### Data is clear

### Overwrite file with clear data set

```python
df.to_csv('full_data.csv', index=False)
```
