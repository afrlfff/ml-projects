# 1. Data collection

##### Import libraires

```python
import pandas as pd
import os
```

##### Make a list of __feature names__ given in "readme" files

```python
feature_names = ['timestamp', 'activityID', 'heart rate']
for body_part in ['hand', 'chest', 'ankle']:
    feature_names += [
        f"{body_part}_temp", 
        f"{body_part}_acc16g_x", f"{body_part}_acc16g_y", f"{body_part}_acc16g_z",
        f"{body_part}_acc6g_x", f"{body_part}_acc6g_y", f"{body_part}_acc6g_z", 
        f"{body_part}_gyro_x", f"{body_part}_gyro_y", f"{body_part}_gyro_z", 
        f"{body_part}_mag_x", f"{body_part}_mag_y", f"{body_part}_mag_z", 
        f"{body_part}_ori_1", f"{body_part}_ori_2", f"{body_part}_ori_3", f"{body_part}_ori_4"
    ]
feature_names
```

##### Merge all data to the single Data Frame

```python
PATH_TO_PROTOCOL = os.path.join(os.getcwd(), 'PAMAP2_Dataset', 'Protocol')
PATH_TO_OPTIONAL = os.path.join(os.getcwd(), 'PAMAP2_Dataset', 'Optional')

df = pd.DataFrame()

for _, _, files in os.walk(PATH_TO_PROTOCOL):
    for file in files:
        if not file.endswith('dat'):
            continue
        temp_df = pd.read_csv(os.path.join(PATH_TO_PROTOCOL, file), names=feature_names, sep='\s+')
        df = pd.concat([df, temp_df])
for _, _, files in os.walk(PATH_TO_OPTIONAL):
    for file in files:
        if not file.endswith('dat'):
            continue
        temp_df = pd.read_csv(os.path.join(PATH_TO_OPTIONAL, file), names=feature_names, sep='\s+')
        df = pd.concat([df, temp_df])

# restore indexes
df = df.reset_index(drop=True)
df.shape
```

##### Save result Data Frame

```python
df.to_csv('full_data.csv', index=False)
```
